# Tratamento da base de dados

**Justificativa:** A base de dados é extensa e possui campos que não irão interferir nos resultados deste estudo.

**Questão do Estudo:** Quais fatores afetam a avaliação da redação do ENEM 2021?

```python
import pandas as pd
```

```python
df_completo = pd.read_csv("/content/drive/MyDrive/MICRODADOS_ENEM_2021.csv", header = 0, delimiter = ';', encoding='latin-1')
```

```python
df_completo.sample(7)
```

```python
df_aluno = df_completo[["NU_INSCRICAO", "TP_FAIXA_ETARIA","TP_SEXO","Q006", "Q025", 
                        "NU_NOTA_REDACAO"]].query("NU_NOTA_REDACAO > 0")
df_escola = df_completo[["NU_INSCRICAO", "TP_DEPENDENCIA_ADM_ESC",
                         "CO_MUNICIPIO_PROVA", "NU_NOTA_REDACAO"]].query("NU_NOTA_REDACAO > 0")
df_escola.drop(["NU_NOTA_REDACAO"], axis=1, inplace=True)
```

```python
df_escola.sample(2)
```

```python
df_aluno.to_csv("/content/drive/MyDrive/ENEM_2021_ALUNO.csv")
df_escola.to_csv("/content/drive/MyDrive/ENEM_2021_ESCOLA.csv")
```
